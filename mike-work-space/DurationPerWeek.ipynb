```python
import glob
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
```

```python
files = glob.glob('data-to-use/*.csv')
```

### Create DF for each month as well as a single dataframe for all data in 2020

```python
df_feb = df_march = df_jan = df_april = df_may = df_2020 = pd.DataFrame(columns= pd.read_csv(files[0]).columns)
for f in files:
    df = pd.read_csv(f)
    df_2020 = df_2020.append(df[(df.years == 2020)],ignore_index= True)
    df_jan = df_jan.append(df[(df.years == 2020) & (df.months == 1)],ignore_index= True)
    df_feb = df_feb.append(df[(df.years == 2020) & (df.months == 2)],ignore_index= True)
    df_march = df_march.append(df[(df.years == 2020) & (df.months == 3)],ignore_index= True)
    df_april = df_april.append(df[(df.years == 2020) & (df.months == 4)],ignore_index= True)
    df_may = df_may.append(df[(df.years == 2020) & (df.months == 5)],ignore_index= True)
```

### Functiuon to plot duration by week number for any passed in DF

```python
def plot_duration_by_week(df):
    font = {'family' : 'normal','weight' : 'bold','size': 16}
    matplotlib.rc('font', **font)
    plt.rcParams["figure.figsize"] = (20,10)
    ax = df.groupby(pd.to_datetime(df['datetime']).dt.strftime('%W'))['durations'].sum().plot()
    ax.set_xlabel("Week #")
    ax.set_ylabel("Total Video Duration (seconds)")
    
```

```python
plot_duration_by_week(df_jan)
```

```python
plot_duration_by_week(df_feb)
```

```python
plot_duration_by_week(df_march)
```

```python
plot_duration_by_week(df_april)
```

```python
plot_duration_by_week(df_may)
```

```python
plot_duration_by_week(df_2020)
```
