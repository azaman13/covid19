```python
text = df_jan['query'].tolist()
map(word_tokenize, text)
df_jan['query'].apply(word_tokenize)
df_jan['query'].apply(word_tokenize).tolist()
tagged_list = nltk.pos_tag_sents(df_jan['query'].apply(word_tokenize).tolist())
tagged_list
```

```python
noun_list = []
for i in tagged_list:
    temp = [j[0] for j in i if j[1].startswith("NN" or "NNP" or "NNS" or "NNPS")]
    noun_list.append(temp)

print(noun_list)
```

```python
adj_list = []
for i in tagged_list:
    temp = [j[0] for j in i if j[1].startswith("JJ" or "JJR" or "JJS")]
    adj_list.append(temp)

print(adj_list)
```

```python
verb_list = []
for i in tagged_list:
    temp = [j[0] for j in i if j[1].startswith("VB" or "VBD" or "VBG" or "VBN" or "VBP" or "VBZ")]
    verb_list.append(temp)

print(verb_list)
```

```python
df_feb = df[(df.months == 1) & (df.years == 2020)]
```

```python
text = df_feb['query'].tolist()
map(word_tokenize, text)
df_feb['query'].apply(word_tokenize)
df_feb['query'].apply(word_tokenize).tolist()
tagged_list = nltk.pos_tag_sents(df_feb['query'].apply(word_tokenize).tolist())
tagged_list
```

```python
noun_list = []
for i in tagged_list:
    temp = [j[0] for j in i if j[1].startswith("NN" or "NNP" or "NNS" or "NNPS")]
    noun_list.append(temp)

print(noun_list)
```

```python
string = ""
for l in noun_list:
    if len(l) > 1:
        for word in l:
            string = string + " " + word
        tokens = nltk.word_tokenize(string)  
        bigrm = nltk.bigrams(tokens) 
        print(*map('-'.join, bigrm), sep=', ')
    else:
        print(l)
    string = ""
```
