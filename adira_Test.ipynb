```python
def tokenize(searches):
    sents = []
    
    for q in searches:
        #tokenize and tag
        tokenlist = word_tokenize(q)
        tagged_list = nltk.pos_tag(tokenlist)
        ##Add sublists to master list of sentiments and searches
        for term in tagged_list: 
            sents.append(term)
            
    return sents
```

```python
def tokenize_only(searches):
    tokens = []
    for q in searches:
        #tokenize and tag
        tokenlist = word_tokenize(q)
        for token in tokenlist:
            tokens.append(token)
```
