```python
import pandas as pd
import matplotlib.pyplot as plt
import dataKey
```

```python
df = pd.read_csv("processed_data.csv")
```

```python
def create_result_histogram(df,field_name):
    try:
        question, answers = dataKey.retrive_question_data(field_name)
    except Exception:
        question = field_name
    df[field_name].value_counts().plot.hist()
    plt.suptitle(question)
```

```python
def create_result_bar(df,field_name):
    dataKey.retrive_question_data(field_name)
    df[field_name].value_counts().plot.bar()
    plt.suptitle(field_name)
```

```python
create_result_histogram(df,"gad7_score")
```

```python
create_result_histogram(df,"selfesteem_score")
```

```python
create_result_histogram(df,"phq9_score")
```

```python
create_result_bar(df,"employment")
```
